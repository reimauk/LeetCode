```python
class Solution:
    def tictactoe(self, moves):
        x = [[0, 0, 0],
             [0, 0, 0],
             [0, 0, 0]]
        g = 1
        for index, move in enumerate(moves):
            x[move[0]][move[1]] = g

            if (x[0][0] == g and x[1][1] == g and x[2][2] == g) or \
                (x[0][2] == g and x[1][1] == g and x[2][0] == g):
                return "A" if g == 1 else "B"

            for i in range(3):
                # Ряды
                if x[i][0] == g and x[i][1] == g and x[i][2] == g:
                    return "A" if g == 1 else "B"
                # Колонки
                if x[0][i] == g and x[1][i] == g and x[2][i] == g:
                    return "A" if g == 1 else "B"

            if index == 8:
                return "Draw"
            g = - g

        return "Pending"
```

```python
prov = Solution()

# Example 1:
moves = [[0,0],[2,0],[1,1],[2,1],[2,2]]
# Output: "A"
print(f"Example 1: {moves}")
print(prov.tictactoe(moves))


# Example 2:
moves = [[0,0],[1,1],[0,1],[0,2],[1,0],[2,0]]
# Output: "B"
print()
print(f"Example 2: {moves}")
print(prov.tictactoe(moves))

# Example 3:
#ops = ["1","C"]
# Output: 0
#print()
#print(f"Example 3: {ops}")
#print(prov.tictactoe(ops))
```
