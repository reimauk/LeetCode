You are given two strings s and t.

String t is generated by random shuffling string s and then add one more letter at a random position.

Return the letter that was added to t.

```python
class Solution:
    def findTheDifference(self, s: str, t: str) -> str:
        """
        :type s: str
        :type t: str
        :rtype: str
        """
        dif = ""
        for i in range(len(t)):
            if t[i] not in s:
                dif += t[i]
            else:
                s = s.replace(t[i], "", 1)
        return dif
```

```python
s = "abcd"
t = "abcde"


prov = Solution()
print(prov.findTheDifference(s, t))
print(prov.findTheDifference(s, t) == "e")

s = "a"
t = "aa"
print(prov.findTheDifference(s, t))
print(prov.findTheDifference(s, t) == "a")
```
