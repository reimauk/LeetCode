Given a square matrix mat, return the sum of the matrix diagonals.

Only include the sum of all the elements on the primary diagonal and all the elements on the secondary diagonal that are not part of the primary diagonal.


```python
class Solution:
    def diagonalSum(self, mat):
        n = len(mat)
        sum1 = 0
        sum2 = 0
        for i in range(n):
            sum1 += mat[i][i]
            if n-i-1 != i:
                sum2 += mat[n-i-1][i]

        return sum1 + sum2
```

```python
class Solution:
    def diagonalSum(self, mat):
        n = len(mat)
        sum = 0
        for i in range(n):
            sum += mat[i][i] + mat[n-i-1][i]
        if n % 2 == 1:
            sum -= mat[n // 2][n // 2]
        return sum
```

```python
prov = Solution()

# Example 1:
mat = [[1,2,3],
       [4,5,6],
       [7,8,9]]
# Output: 25
print(mat)
print(prov.diagonalSum(mat))
```
