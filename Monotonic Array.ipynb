An array is monotonic if it is either monotone increasing or monotone decreasing.

An array nums is monotone increasing if for all i <= j, nums[i] <= nums[j]. An array nums is monotone decreasing if for all i <= j, nums[i] >= nums[j].

Given an integer array nums, return true if the given array is monotonic, or false otherwise.

```python
class Solution:
    def isMonotonic(self, nums):
        monotonic = 0
        monotonic_next = 0
        for i in range(1, len(nums)):
            dif = nums[i] - nums[i-1]
            if dif > 0:
                monotonic_next = 1
            elif dif < 0:
                monotonic_next = -1
            if monotonic == 1 and monotonic_next == -1:
                return False
            if monotonic == -1 and monotonic_next == 1:
                return False
            #print(f"dif={dif}, m={monotonic}, m2={monotonic_next}")
            if monotonic_next == 1 or monotonic_next == -1:
                monotonic = monotonic_next
        return True
```

```python
class Solution:
    def isMonotonic(self, nums):
        decreasing = True
        increasing = True
        for i in range(1, len(nums)):
            dif = nums[i] - nums[i - 1]
            if dif > 0:
                decreasing = False
            if dif < 0:
                increasing = False

            if not decreasing and not increasing:
                return False
        return True
```

Example 1:

Input: nums = [1,2,2,3]
Output: true
Example 2:

Input: nums = [6,5,4,4]
Output: true
Example 3:

Input: nums = [1,3,2]
Output: false
