Roman numerals are represented by seven different symbols: I, V, X, L, C, D and M.

Symbol       Value
I             1
V             5
X             10
L             50
C             100
D             500
M             1000
For example, 2 is written as II in Roman numeral, just two ones added together. 12 is written as XII, which is simply X + II. The number 27 is written as XXVII, which is XX + V + II.

Roman numerals are usually written largest to smallest from left to right. However, the numeral for four is not IIII. Instead, the number four is written as IV. Because the one is before the five we subtract it making four. The same principle applies to the number nine, which is written as IX. There are six instances where subtraction is used:

I can be placed before V (5) and X (10) to make 4 and 9.
X can be placed before L (50) and C (100) to make 40 and 90.
C can be placed before D (500) and M (1000) to make 400 and 900.
Given a roman numeral, convert it to an integer

```python
class Solution:
    def romanToInt(self, s):
        length = len(s)
        index = 0
        number = 0
        while index < length:
            if len(s[index:]) > 1:
                symbol = s[index:index + 2]
                if symbol in ["CM", "CD", "XC", "XL", "IX", "IV"]:
                    if symbol == "CM":
                        number += 900
                    if symbol == "CD":
                        number += 400
                    if symbol == "XC":
                        number += 90
                    if symbol == "XL":
                        number += 40
                    if symbol == "IX":
                        number += 9
                    if symbol == "IV":
                        number += 4
                    index += 2
                    continue
            symbol = s[index:index + 1]
            if symbol == "M":
                    number += 1000
            if symbol == "D":
                    number += 500
            if symbol == "C":
                    number += 100
            if symbol == "L":
                    number += 50
            if symbol == "X":
                    number += 10
            if symbol == "V":
                    number += 5
            if symbol == "I":
                    number += 1
            index += 1
        return number
```

```python
class Solution:
    def romanToInt(self, s):
        roman_number2 = {"CM":900, "CD":400, "XC":90, "XL":40, "IX":9, "IV":4}
        roman_number = {"M":1000, "D":500, "C":100, "L":50, "X":10, "V":5, "I":1}
        length = len(s)
        index = 0
        number = 0
        while index < length:
            if len(s[index:]) > 1:
                symbol = s[index:index + 2]
                if symbol in roman_number2:
                    number += roman_number2[symbol]
                    index += 2
                    continue
            symbol = s[index:index + 1]
            number += roman_number[symbol]
            index += 1
        return number
```

```python
class Solution:
    def romanToInt(self, s):
        roman_number = {"MM":2000, "CM":900, "CD":400, "CC":200, "XC":90, "XL":40, "XX":20, "IX":9, "IV":4, "II":2,
                        "M":1000, "D":500, "C":100, "L":50, "X":10, "V":5, "I":1}
        length = len(s)
        index = 0
        number = 0
        while index < length:
            if len(s[index:]) > 1:
                symbol = s[index:index + 2]
                if symbol in roman_number:
                    number += roman_number[symbol]
                    index += 2
                    continue
            number += roman_number[s[index:index + 1]]
            index += 1
        return number
```

Example 1:

Input: s = "III"
Output: 3
Explanation: III = 3.
Example 2:

Input: s = "LVIII"
Output: 58
Explanation: L = 50, V= 5, III = 3.
Example 3:

Input: s = "MCMXCIV"
Output: 1994
Explanation: M = 1000, CM = 900, XC = 90 and IV = 4.


Constraints:

1 <= s.length <= 15
s contains only the characters ('I', 'V', 'X', 'L', 'C', 'D', 'M').
It is guaranteed that s is a valid roman numeral in the range [1, 3999].
