Given a string s, check if it can be constructed by taking a substring of it and appending multiple copies of the substring together

```python
class Solution:
    def repeatedSubstringPattern(self, s: str) -> bool:
        """
        :type s: str
        :rtype: bool
        """
        result = False
        len_string = len(s)
        # Iterate from 1 to length of string s // 2
        for len_substring in range(1, round(len_string // 2) + 1): # "+1" because we started from 1, but not 0
            if len_string % len_substring == 0:
                n = len_string // len_substring # Number repetitions
                index = 0 # starting index in substring
                while index + len_substring < len_string:
                    if s[index : index + len_substring] * n == s:
                        result = True
                        break
                    index += 1
                if result: # If found option "Ok" than break loop
                    break

        return result
```

```python
class Solution:
    def repeatedSubstringPattern(self, s: str) -> bool:
        """
        :type s: str
        :rtype: bool
        """
        result = False
        len_string = len(s)
        # Iterate from 1 to length of string s // 2
        for len_substring in range(1, round(len_string // 2) + 1): # "+1" because we started from 1, but not 0
            if len_string % len_substring == 0:
                n = len_string // len_substring # Number repetitions
                if s[: len_substring] * n == s:
                    return True
        return False
```
